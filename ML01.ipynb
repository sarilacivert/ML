## ML01 Linear Regression

### Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline 
```

#### Import the dataset and extracting the dependent and independent variables

```python
#Importing

salary_data = pd.read_csv('salaries.csv')
list(salary_data.columns)
```

```python
#Assigning Variables

X = salary_data.iloc[:, :-1].values # independent variable

y = salary_data.iloc[:, 1].values # dependent variable
```

```python
y
```

#### Visualizing the dataset

```python
sns.displot(salary_data['YearsExperience'], kde=False, bins=10)
```

```python
sns.countplot(y ='YearsExperience', data=salary_data) 
```

```python
sns.barplot(x = 'YearsExperience', y = 'Salary', data = salary_data)
```

```python
sns.heatmap(salary_data.corr())
```

### Splitting the dataset into the Training Set and Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state=0)
```

#### Fitting Simple Linear Regression to the Training set

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

#### Predicting the Test set results 

```python
y_pred = lr.predict(X_test)
y_pred
```

### Visualising the Training set results

```python
plt.scatter(X_train, y_train, color = 'blue')
plt.plot(X_train, lr.predict(X_train), color = 'red')
plt.title('Salary ~ Experience (Train set)')
plt.xlabel ('Years of Experience')
plt.ylabel('Salary')
plt.show()            
```

### Visualising the Test set results

```python
plt.scatter(X_test, y_test, color = 'blue')
plt.plot(X_train, lr.predict(X_train), color = 'red')
plt.title('Salary ~ Experience (Test set)')
plt.xlabel ('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

### Finding the residuals to measure the accuracy

```python
# Calculating the Residuals

from sklearn import metrics

print ('MAE', metrics.mean_absolute_error(y_test, y_pred), '\n')
print ('MSE', metrics.mean_squared_error(y_test, y_pred), '\n')
print ('RMSE', np.sqrt(metrics.mean_absolute_error(y_test, y_pred)))
```

### the lower the better for the above accuracy scores
